```python
import numpy as np
import scipy.stats as st
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

В качестве распределения было выбрано экспоненциальное. https://en.wikipedia.org/wiki/Exponential_distribution . Согласно википедии функция плотности определяется по след формуле:
\begin{align}
f(x;\lambda) = \begin{cases}
\lambda e^{-\lambda x} & x \ge 0, \\
0 & x < 0.
\end{cases}
\end{align}

Математическое ожидание и дисперсия определяются по формулам:

$$ E = 1/\lambda $$
$$ Var = 1/\lambda^2 $$

Таким образом функция распределения, мат ожидание и дисперсия зависят только от значения лямбды.
Согласно документации https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon в модуле stats для плотности экспоненциального распределения входным параметром является scale, который определяется как:
$$ scale = 1/\lambda $$
Для реализации ЦПТ лямбда было выбрано равным 2.

```python
# значение лямбда
lmbd = 2
# расчет параметра scale
scale = 1/lmbd
# распределение с заданным лямбда
distr = st.expon(scale = scale)
#расчет мат ожидаия и дисперсии
E = 1/lmbd
var = 1/lmbd**2
```

Генерируем выборку объемом 1000 экземпляров из созданного распределения distr

```python
n = 1000 # количество элементов выборки
sample = distr.rvs(n) # генерация 
plt.figure(figsize = (13,10))
plt.hist(sample, density= True, label = u'Гистограмма выборки')
#значения для графика теоретического распределения от 0 до максимального из сгенерированных экземпляров distr
x = np.arange(0,max(sample),0.1) 
plt.plot(x,distr.pdf(x), label = u'Теоретическое распределение')
plt.legend()
plt.xlabel('X')
plt.ylabel(u'Плотность вероятности')
plt.title(u'График выборочного и теоретического распределения')
plt.show()
```

Функция для отрисовки графиков выборочного среднего и теоретического нормального распределения с параметрами из распределения distr. 
На вход в функцию подаются:
1. distrib - сгенерированное распределение
2. n - объем выборки, который будет генерироваться из distr
3. E - математическое ожидание распределения distrib
4. Var - дисперсия распределения distrib

```python
def CPT (distrib, n, E, var):
    # генерация 1000 выборок объема n
    sample = distrib.rvs(size = (1000, n))
    # расчет средних значений каждой выборки
    sample_mean = sample.mean(axis = 1)
    # матожидание и дисперсия нормального распределения на основе данных заданного распределения distrib
    E_norm = E
    var_norm = var/n
    # генерация нормального распределения с параметрами матожидания и дисперсии
    rv_norm = st.norm(E_norm, var_norm**0.5)
    plt.figure(figsize = (13,10))
    #значения для графика теоретического нормального распределения от +- 3 сигма от среднего для построения графика
    x = np.arange(E_norm - 3*var_norm**0.5, E_norm + 3*var_norm**0.5, 0.001)
    #построение гистограммы распределения distrib по 1000 средним выборок и теоретического нормального распределения
    plt.hist(sample_mean, bins= 50,  density= True, label = u'Гистограмма выборки')
    plt.plot(x,rv_norm.pdf(x), label =u'Теоретическое нормальное распределение')
    plt.legend()
    plt.xlabel('X')
    plt.ylabel(u'Плотность вероятности')
    plt.title(u'График выборочного и теоретического распределения для выборки объема n =' + str(n))
    plt.show()
```

```python
n = 5
CPT (distr, n, E, var)
```

```python
n = 10
CPT (distr, n, E, var)
```

```python
n = 50
CPT (distr, n, E, var)
```

Для приведенного экспоненциального распределения с увеличением числа n меняется вид гистограммы распределения. Чем меньше n, тем лучше видна асимметричность (правая часть распределения имеет более длинный хвост), соответственно с ростом n ассиметричность уменьшается.
