```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Логнормальное распределение



<a href="https://en.wikipedia.org/wiki/Log-normal_distribution"> wiki </a>

Плотность вероятности: ${\displaystyle \exp \left(-\left.\left[{\frac {\ln(x)-\mu }{\sigma }}\right]^{2}\right/2\right)\left/\left(x\sigma {\sqrt {2\pi }}\right)\right.}$<br>
Функция распределения: ${\displaystyle {\frac {1}{2}}+{\frac {1}{2}}\mathrm {Erf} \left[{\frac {\ln(x)-\mu }{\sigma {\sqrt {2}}}}\right]}$<br>
Математическое ожидание: ${\displaystyle e^{\mu +\sigma ^{2}/2}}$<br>
Медиана:	${\displaystyle e^{\mu }}$<br>
Мода:	${\displaystyle e^{\mu -\sigma ^{2}}}$<br>
Дисперсия:	${\displaystyle (e^{\sigma ^{2}}\!\!-1)e^{2\mu +\sigma ^{2}}}$

Пусть: ${\mu = 0}$, а ${\sigma = 1}$<br>

```python
m = 0
s = 1 
# зададим случайную величину
lognorm_rv = sts.lognorm(s)

# выборка
sample = lognorm_rv.rvs(size=200)
```

Теоретическая функция распределения и плотность вероятности:

```python
x = np.linspace(0,5,100)
cdf = lognorm_rv.cdf(x)
pdf = lognorm_rv.pdf(x)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
axes[0].plot(x, cdf, 'r')
axes[1].plot(x, pdf, 'g')
axes[0].set_xlabel("$x$")
axes[1].set_xlabel("$x$")
axes[0].set_ylabel("CDF")
axes[1].set_ylabel("PDF")
axes[0].set_title('Теоретическая функция распределения')
axes[1].set_title('Теоретическая плотность вероятности')
plt.show()

```

Гистограмма выборки:

```python
plt.figure(figsize=(13,9))
plt.hist(sample, density=True, bins =30, histtype='stepfilled', alpha=0.8, label = 'Fraction of samples')
x = np.linspace(0,25, 1000)
pdf = lognorm_rv.pdf(x)
plt.plot(x, pdf, 'r--', label='Theoretical PDF', lw = 2)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend(loc='upper center')
plt.show()
```

Математическое ожидание и дисперсия:

```python
M = math.exp(m + s**2 / 2)
D = (math.exp(s**2) - 1) * math.exp(2 * m + s**2)
print("Mean:", M, "Variance:", D)

```

Функция для построения гистограммы 1000 средних и плотности распределения нормального распределения для заданного n:

```python
def CPT(n):

    msample = lognorm_rv.rvs(size=(1000,n)) # 1000 выборок по n значений
    mlist = msample.mean(axis=1) # 1000 значений, где каждое - среднее из строки выборки
    norm_rv = sts.norm(M, (D / n) ** 0.5) # функция нормального распределения
    x = np.linspace(min(mlist) - 0.5,max(mlist) + 0.5, 1000)
    plt.figure(figsize=(10,7))

    plt.hist(mlist, density=True, bins =30, histtype='stepfilled', alpha=0.8, label = 'Fraction of means. n=' + str(n))
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, 'r--', label='Normal PDF. n=' + str(n), lw = 2)
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.legend(loc='best')
    plt.show()

```

```python
CPT(5) # Проверка ЦПТ для размера выборок n = 5
```

```python
CPT(10) # Проверка ЦПТ для размера выборок n = 10
```

```python
CPT(50) # Проверка ЦПТ для размера выборок n = 50
```

```python
CPT(1000) # Проверка ЦПТ для размера выборок n = 1000
```
